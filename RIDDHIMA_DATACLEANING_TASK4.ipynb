DROPPING DUPLICATE ROWS (IF ANY)

```python
data=data.drop_duplicates()
data
```

```python
print(data.basement.unique())
print(data.mainroad.unique())
print(data.guestroom.unique())
print(data.prefarea.unique())
```

PUTTING ALL UNIQUES OF YES/NO IN A LIST TO REPLACE THEM

```python
stringsy=['Haanji','True','Indeed','y','Yo','yep','yes','Haan','Of course','Y',"['y']"]
repy='YES'
stringsn=['no','Nakko','False','Nah','n','N','Naaji','Nope','ille',"['n']"]
repn='NO'
```

REPLACING ALL Y,YEP,HAANJI etc with YES AND n.NAHH,NAAJI etc with NO

```python
#putting yes
data['basement']=data['basement'].replace(stringsy,repy)
data["mainroad"]=data["mainroad"].replace(stringsy,repy)
data["guestroom"]=data["guestroom"].replace(stringsy,repy)
data["hotwaterheating"]=data["hotwaterheating"].replace(stringsy,repy)
data["prefarea"]=data["prefarea"].replace(stringsy,repy)
data["airconditioning"]=data["airconditioning"].replace(stringsy,repy)
#putting no
data['basement']=data['basement'].replace(stringsn,repn)
data["mainroad"]=data["mainroad"].replace(stringsn,repn)
data["guestroom"]=data["guestroom"].replace(stringsn,repn)
data["hotwaterheating"]=data["hotwaterheating"].replace(stringsn,repn)
data["prefarea"]=data["prefarea"].replace(stringsn,repn)
data["airconditioning"]=data["airconditioning"].replace(stringsn,repn)
data
```

REMOVING COLUMNS HAVING THRESHOLD OF 8 NAN VALUES

```python
data=data.dropna(axis=0,thresh=9)
data = data.reset_index(drop=True)
data.shape
```

USING INTERPOLATE TO FILL NAN VALUES IN THE COLUMN AREA

```python
data['area']=data['area'].interpolate(method='linear').fillna(1).astype(int)
data
```

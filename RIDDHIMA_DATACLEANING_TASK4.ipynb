```python
import pandas as pd
import numpy as np
```

IMPORTING CSV FILE AS DATAFRAME

```python
data=pd.read_csv("Task_4.csv")
```

```python
data
```

DROPPING DUBLICATEROWS (IF ANY)

```python
data=data.drop_duplicates()
data
```

PRINTS TOTAL NAN VALUES IN THE DATASET

```python
data.isnull().sum()
data.isnull().sum().sum()
```

PRINTS UNIQUE VALUES OF COLUMNS BASEMENT,MAINROAD,GUESTROOM

```python
print(data.basement.unique())
print(data.mainroad.unique())
print(data.guestroom.unique())
```

PUTTING ALL UNIQUES OF YES/NO IN A LIST TO REPLACE THEM

```python
stringsy=['Haanji','True','Indeed','y','Yo','yep','yes','Haan','Of course','Y',"['y']"]
repy='YES'
stringsn=['no','Nakko','False','Nah','n','N','Naaji','Nope','ille',"['n']"]
repn='NO'
```

REPLACING ALL Y,YEP,HAANJI etc with YES AND n.NAHH,NAAJI etc with NO

```python
data['basement']=data['basement'].replace(stringsy,repy)
data["mainroad"]=data["mainroad"].replace(stringsy,repy)
data["guestroom"]=data["guestroom"].replace(stringsy,repy)
data["hotwaterheating"]=data["hotwaterheating"].replace(stringsy,repy)
data["prefarea"]=data["prefarea"].replace(stringsy,repy)
data["airconditioning"]=data["airconditioning"].replace(stringsy,repy)
data['basement']=data['basement'].replace(stringsn,repn)
data["mainroad"]=data["mainroad"].replace(stringsn,repn)
data["guestroom"]=data["guestroom"].replace(stringsn,repn)
data["hotwaterheating"]=data["hotwaterheating"].replace(stringsn,repn)
data["prefarea"]=data["prefarea"].replace(stringsn,repn)
data["airconditioning"]=data["airconditioning"].replace(stringsn,repn)
data
```

REMOVING COLUMNS HAVING THRESHOLD OF 8 NAN VALUES

```python
data=data.dropna(axis=0,thresh=8)
data = data.reset_index(drop=True)
```

USING INTERPOLATE TO FILL NAN VALUES IN THE COLUMN AREA

```python
data['area']=data['area'].interpolate(method='linear').fillna(-1).astype(int)
data
```

DISPLAY.MAX_ROWS HELPS DISPLAY ALL ROWS INSTEAD OF FIRST 5 AND LAST 5

```python
pd.set_option('display.max_rows', None)
data
```

PUTTING no in place of nan values in a column having unfurnished as furnishing status

```python
data = data.apply(lambda row: row.fillna('NO') if 'unfurnished' in row.values else row, axis=1)
data
```

```python
data['bedrooms']=data['bedrooms'].replace('NO',np.nan)
data['bathrooms']=data['bathrooms'].replace('NO',np.nan)
data['stories']=data['stories'].replace('NO',np.nan)
data['parking']=data['parking'].replace('NO',np.nan)
data
```

```python
data['bedrooms'] = data['bedrooms'].interpolate(method='linear').fillna(2).astype(int)
data
```

```python
data['bathrooms'].fillna(data['bedrooms'] - 1, inplace=True)
data
```

```python
mean_value = int(data['stories'].mean())
data['stories'].fillna(mean_value, inplace=True)
data
```

```python
data['mainroad'].fillna('YES',inplace=True)
data['hotwaterheating'].fillna('NO',inplace=True)
data
```

```python
mean = data['parking'].mean()
data['parking'].fillna(1, inplace=True)
data
```

```python
data['airconditioning'].fillna(method='ffill',inplace=True)
data
```

```python
data['guestroom'] = data['price'].apply(lambda x: 'YES' if 2000 <= x <= 7000 else 'NO')
data
```

```python
yes = data['furnishingstatus'].eq('YES').sum()
no= data['furnishingstatus'].eq('NO').sum()
if yes >3:
    data['furnishingstatus'].fillna('furnished',inplace=True)
elif no>3:
    data['furnishingstatus'].fillna('unfurnished',inplace=True)
else:
    data['furnishingstatus'].fillna('semi-furnished',inplace=True)
data
```
